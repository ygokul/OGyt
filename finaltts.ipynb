```python
from translate import Translator
from gtts import gTTS
import os
import platform
import textwrap
from yt_dlp import YoutubeDL
import assemblyai as aai
def download_audio(url, output_filename):
    try:
        # Ensure the filename has the .mp3 extension
        if not output_filename.endswith('.mp3'):

            ydl_opts = {
                'format': 'bestaudio/best',
                'postprocessors': [{
                    'key': 'FFmpegExtractAudio',
                    'preferredcodec': 'mp3',
                    'preferredquality': '192',
                }],
                'outtmpl': output_filename  # Save as the specified filename
            }
        
        with YoutubeDL(ydl_opts) as ydl:
            ydl.download([url])
        return output_filename+".mp3"
    except Exception as e:
        print("Error:", e)
        return None

def browse_path_or_url():
    while True:
        user_input = input("Enter the URL of the video (or 'exit' to quit): ")
        if user_input.lower() == "exit":
            return None
        elif "http" in user_input:
            return user_input
        else:
            print("Invalid input. Please enter a valid URL.")

def transcribe_audio(file_path):
    try:
        aai.settings.api_key = "8ded3e61ee47493b8d2bcc093ca1bf67"
        transcriber = aai.Transcriber()
        transcript = transcriber.transcribe(file_path)
        return transcript.text
    except Exception as e:
        print("Error during transcription:", e)
        return None

if __name__ == "__main__":
    print("Please enter the URL of the video.")
    print("Type 'exit' to quit.")
    video_input = browse_path_or_url()

    if not video_input:
        print("No input provided. Exiting.")
        exit()

    output_filename = "trans"
    audio_path = download_audio(video_input, output_filename)

    if not audio_path:
        print("Failed to download the audio. Exiting.")
        exit()
    
    print("Audio downloaded successfully at:", audio_path)

    # Transcribe the downloaded audio file
    transcription = transcribe_audio(audio_path)

    if transcription:
        print("Transcription successful,English text:" +transcription)
    else:
        print("Transcription failed.")

available_languages = {
    'Tamil': 'ta',
    'Hindi': 'hi',
    'French': 'fr',
    'Spanish': 'es',
    'German': 'de',
    'Chinese': 'zh-CN',
    'Japanese': 'ja'
}
def translate_text_to_audio(text, target_lang):
    # Step 1: Split text into smaller chunks
    chunk_size = 500  # Limit of characters per request
    chunks = textwrap.wrap(text, chunk_size)
    translated_text = ""

    # Step 2: Translate each chunk and combine results
    translator = Translator(to_lang=target_lang)
    for chunk in chunks:
        translated_chunk = translator.translate(chunk)
        translated_text += translated_chunk + " "
    
    # Step 3: Convert the translated text to speech
    tts = gTTS(text=translated_text, lang=target_lang)
    audio_file = f"translated_audio_{target_lang}.mp3"
    tts.save(audio_file)
    
    # Print the translated text for reference
    print("Translated Text:", translated_text)
    
    # Step 4: Play the audio based on the operating system
    if platform.system() == "Windows":
        os.system(f"start {audio_file}")
    elif platform.system() == "Darwin":  # macOS
        os.system(f"afplay {audio_file}")
    else:  # Linux
        os.system(f"mpg321 {audio_file}")

# Example usage
english_text = transcription  # Replace with transcript.text or any English text

# Prompt user to select a language
print("Available languages:")
for lang_name in available_languages:
    print(f"- {lang_name}")

selected_lang_name = input("Please enter your target language: ")
target_lang = available_languages.get(selected_lang_name.capitalize())

if target_lang:
    translate_text_to_audio(english_text, target_lang)
else:
    print("Invalid language selection.")
```
